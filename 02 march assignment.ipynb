```python
Matplotlib is a Python library used for creating static, animated, and interactive visualizations in Python. It is one of the
most popular data visualization libraries in Python and can be used to create a wide range of visualizations including line 
plots, scatter plots, histograms, bar plots, and many more.
Matplotlib is used to create high-quality and publication-ready visualizations that can be used for data exploration, 
presentation, and communication. It is also a popular choice for scientific visualization and is used extensively in the 
scientific community.
Pyplot is a sub-library of Matplotlib that provides a simple interface for creating visualizations. Some of the popular plots
that can be created using the Pyplot module of Matplotlib include:
1.Line plots
2.Scatter plots
3.Bar plots
4.Histograms
5.Pie charts
```

```python
Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
```

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel("this is x data")
plt.ylabel("this is y data")
plt.title("data")
plt.show()
```

```python
Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the plots
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Creating the subplots
fig, axs = plt.subplots(2, 2)

# Plotting the data on the subplots
axs[0, 0].plot(x1, y1)
axs[0, 1].plot(x2, y2)
axs[1, 0].plot(x3, y3)
axs[1, 1].plot(x4, y4)

# Adding titles to the subplots
axs[0, 0].set_title('Line 1')
axs[0, 1].set_title('Line 2')
axs[1, 0].set_title('Line 3')
axs[1, 1].set_title('Line 4')

# Displaying the plot
plt.show()
```

```python
Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
```

```python
A bar plot is a type of plot in which rectangular bars are used to represent the values of a categorical variable or a set of
data. The length or height of each bar is proportional to the value that it represents.
Bar plots are used to visualize and compare the numerical values of different categories or groups. They are particularly 
useful for visualizing discrete or categorical data, where each category represents a distinct group or entity.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
plt.show()
```

```python
Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
```

```python
A box plot, also known as a box-and-whisker plot, is a type of plot that displays the distribution of a set of data by showing
the median, quartiles, and range of the data. The box in the plot represents the interquartile range (IQR), which is the range
of values between the first and third quartiles (25th and 75th percentiles), while the whiskers represent the range of values
outside of the IQR.
Box plots are used to summarize and compare the distribution of data for different groups or variables. They are particularly
useful for detecting outliers and visualizing the spread and skewness of data.
import matplotlib.pyplot as plt
import numpy as np

# Data for the box plot
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1, box2]

# Creating the box plot
fig, ax = plt.subplots()
ax.boxplot(data)

# Adding labels and title to the plot
ax.set_xticklabels(['Box 1', 'Box 2'])
ax.set_ylabel('Value')
ax.set_title('Box Plot Example')

# Displaying the plot
plt.show()
```
