```python
Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would
use each type of test.
```

```python
  Both t-test and z-test are statistical hypothesis tests used to determine whether two populations have different means. The
key difference between the two is in the calculation of the test statistic and the assumption about the population variance.

The z-test assumes that the population variance is known, or the sample size is large enough that the sample variance is a 
good estimate of the population variance. The test statistic is calculated as the difference between the sample mean and the 
hypothesized population mean, divided by the standard error of the mean. The resulting z-score is compared to a critical value
from the standard normal distribution to determine the p-value and make a decision about the null hypothesis.

A t-test, on the other hand, is used when the population variance is unknown and needs to be estimated from the sample data. 
The test statistic is calculated as the difference between the sample mean and the hypothesized population mean, divided by
the standard error of the mean estimated from the sample data. The resulting t-score is compared to a critical value from the 
t-distribution with n-1 degrees of freedom to determine the p-value and make a decision about the null hypothesis.

An example scenario where a z-test might be used is when testing whether the mean weight of a certain population is different
from a hypothesized value. If the population standard deviation is known or the sample size is large enough, a z-test can be
used to make this determination.

An example scenario where a t-test might be used is when testing whether the mean score on a certain exam is different between
two groups of students. If the standard deviation of the scores is unknown, a t-test can be used to estimate the standard 
error of the mean and make this determination.
```

```python
Q2: Differentiate between one-tailed and two-tailed tests
```

```python
 A hypothesis test is a statistical procedure used to determine whether a hypothesis about a population parameter is supported
by sample data. The null hypothesis (H0) is the hypothesis that there is no significant difference between a population 
parameter and a hypothesized value, while the alternative hypothesis (Ha) is the hypothesis that there is a significant 
difference.

One-tailed and two-tailed tests refer to the directionality of the alternative hypothesis. In a one-tailed test, the 
alternative hypothesis specifies the direction of the difference between the sample statistic and the hypothesized value. For 
example, the alternative hypothesis for a one-tailed test of a mean might be that the population mean is greater than the 
hypothesized value. In a two-tailed test, the alternative hypothesis specifies that the difference between the sample 
statistic and the hypothesized value could be in either direction. For example, the alternative hypothesis for a two-tailed 
test of a mean might be that the population mean is not equal to the hypothesized value.

The choice between a one-tailed and two-tailed test depends on the research question and the directionality of the effect 
being tested. A one-tailed test is appropriate when the research question is specifically focused on a particular direction of
the effect, and the null hypothesis is unlikely to be rejected if the effect is in the opposite direction. A two-tailed test
is appropriate when the research question is open-ended and the null hypothesis could be rejected regardless of the direction
of the effect.

In general, a one-tailed test has greater power to detect an effect in the specified direction than a two-tailed test, but 
also has a greater risk of a type I error (false positive) if the effect is actually in the opposite direction.
```

```python
Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for
each type of error.
```

```python
  Type 1 and Type 2 errors are two types of errors that can occur in hypothesis testing.

Type 1 error is the rejection of a true null hypothesis, which means that the statistical test detects an effect that is not
really there. Type 1 error is also called a false positive. The probability of making a Type 1 error is denoted by alpha (α), 
and is typically set at 0.05 or 0.01.

   Example scenario: A pharmaceutical company is testing a new drug to lower blood pressure. The null hypothesis is that the 
drug has no effect on blood pressure. If the company incorrectly rejects the null hypothesis and concludes that the drug is 
effective when it is actually not, this is a Type 1 error.

Type 2 error is the failure to reject a false null hypothesis, which means that the statistical test fails to detect an effect
that is actually there. Type 2 error is also called a false negative. The probability of making a Type 2 error is denoted by 
beta (β), and is typically set at 0.2 or 0.1.

   Example scenario: A medical researcher is testing a new cancer treatment. The null hypothesis is that the treatment has no 
    effect on cancer progression. If the researcher fails to reject the null hypothesis and concludes that the treatment is not
    effective when it is actually effective, this is a Type 2 error.

The choice between Type 1 and Type 2 errors depends on the research question and the consequences of each error. In some cases,
a Type 1 error may be more serious than a Type 2 error (such as in a medical diagnosis), while in other cases, a Type 2 error 
may be more serious than a Type 1 error (such as in a quality control test).
```

```python
  Bayes's theorem is a mathematical formula used for calculating conditional probabilities. It provides a way of updating 
probabilities based on new evidence or information.

The theorem can be stated as follows:

    P(A|B) = P(B|A) * P(A) / P(B)

where P(A|B) is the probability of A given B, P(B|A) is the probability of B given A, P(A) is the prior probability of A, and 
P(B) is the prior probability of B.

Here is an example to illustrate Bayes's theorem:

Suppose there is a medical test for a certain disease that is 95% accurate. That is, the test correctly identifies 95% of 
people who have the disease (true positives) and 95% of people who do not have the disease (true negatives). Let's assume that 
the prevalence of the disease in the population is 1 in 1000.

Now, suppose a person takes the test and the result comes back positive. What is the probability that the person actually has 
the disease?

Using Bayes's theorem, we can calculate this probability as follows:

Let A be the event that the person has the disease, and let B be the event that the test is positive.

    P(A) = 1/1000 (prior probability of the disease in the population)
    P(B|A) = 0.95 (probability of a positive test result given the person has the disease)
    P(B|not A) = 0.05 (probability of a positive test result given the person does not have the disease)

We can use the complement rule to calculate P(not A):

    P(not A) = 1 - P(A) = 999/1000

Now we can use Bayes's theorem to calculate P(A|B), the probability that the person has the disease given a positive test 
result:

    P(A|B) = P(B|A) * P(A) / [P(B|A) * P(A) + P(B|not A) * P(not A)]

    P(A|B) = 0.95 * 1/1000 / [0.95 * 1/1000 + 0.05 * 999/1000]

    P(A|B) = 0.0185

So, the probability that the person actually has the disease given a positive test result is only 1.85%. This example 
demonstrates the importance of considering both the sensitivity and specificity of a test, as well as the prevalence of the 
disease in the population, when interpreting test results.
```

```python
Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.
```

```python
  A confidence interval is a range of values that is likely to contain the true value of a population parameter with a certain level of 
    confidence. It is calculated from a sample of data, and the width of the interval depends on the sample size and the 
    desired level of confidence.

To calculate the confidence interval, we first determine the point estimate of the population parameter (such as the mean or
proportion) from the sample data. We then use the standard error formula to estimate the variability of the sampling 
distribution. Finally, we use a t-distribution or z-distribution, depending on the sample size and whether the population 
standard deviation is known, to determine the critical value for the desired level of confidence. The confidence interval is 
then calculated as:

Confidence Interval = Point Estimate ± (Critical Value × Standard Error)

For example, suppose we want to estimate the mean height of all adult males in a certain city. We take a random sample of 50 
adult males and find that the sample mean height is 68 inches with a sample standard deviation of 2 inches. We want to 
calculate a 95% confidence interval for the population mean height.

First, we determine the point estimate of the population mean height, which is the sample mean height of 68 inches. Then, we
calculate the standard error using the formula:

Standard Error = Sample Standard Deviation / Square Root of Sample Size
Standard Error = 2 / √50
Standard Error = 0.28

Next, we use a t-distribution with 49 degrees of freedom (50-1) to determine the critical value for a 95% confidence interval.
From a t-table, we find the critical value to be 2.009.

Finally, we can calculate the confidence interval as:

Confidence Interval = 68 ± (2.009 × 0.28)
Confidence Interval = (67.44, 68.56)

Therefore, we can say with 95% confidence that the true population mean height of adult males in the city is between 67.44 and
68.56 inches.
```

```python
Q6. Use Bayes' Theorem to calculate the probability of an event occurring given prior knowledge of the
event's probability and new evidence. Provide a sample problem and solution.
```

```python
 Problem:
 
 Suppose a company produces two different types of electronic components, A and B. Component A has a defect rate of 5%, while
    component B has a defect rate of 10%. A customer reports receiving a defective component, but it is not known whether it
    was type A or type B. What is the probability that the defective component was type A?

Solution:
    
Let D be the event that the component is defective, and let A and B be the events that the component is type A or type B, 
respectively. We want to find P(A|D), the probability that the component is type A given that it is defective.

By Bayes' Theorem, we have:

    P(A|D) = P(D|A) * P(A) / P(D)

where P(D|A) is the probability that the component is defective given that it is type A, P(A) is the prior probability that 
the component is type A, and P(D) is the marginal probability of the component being defective.

We know that P(D|A) = 0.05, since type A components have a 5% defect rate. We also know that P(A) = 0.5, since there are two 
types of components and they are equally likely to be selected. Finally, we can use the law of total probability to find P(D):

    P(D) = P(D|A) * P(A) + P(D|B) * P(B)

where P(D|B) is the probability that the component is defective given that it is type B, and P(B) is the prior probability 
that the component is type B. We know that P(D|B) = 0.1, and P(B) = 0.5, so we can substitute these values:

    P(D) = 0.05 * 0.5 + 0.1 * 0.5 = 0.075

Now we can plug all these values into Bayes' Theorem:

    P(A|D) = 0.05 * 0.5 / 0.075 = 0.333

Therefore, the probability that the defective component was type A is 0.333, or approximately 33.3%.
```

```python
  To calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation of 5, we need to 
    use the following formula:

    95% confidence interval = sample mean ± 1.96 x standard error

where the standard error is calculated as:

    standard error = standard deviation / square root of sample size

Assuming a sample size of 100, we can calculate the standard error as:

    standard error = 5 / sqrt(100) = 0.5

Substituting the values in the formula, we get:

    95% confidence interval = 50 ± 1.96 x 0.5 = (49.01, 50.99)

Interpretation: We can say with 95% confidence that the true population mean lies between 49.01 and 50.99.
```

```python
 The margin of error is the maximum expected difference between the true population parameter and the estimated sample 
    statistic. It is a measure of the precision of the estimate and is typically expressed as a percentage. The margin of error
    is calculated by multiplying the critical value of the confidence interval by the standard error of the sample mean.

Sample size has a direct effect on the margin of error. As the sample size increases, the margin of error decreases. This is 
because a larger sample size results in a more representative sample of the population, which in turn reduces the variability
of the sample mean and increases the precision of the estimate.

For example, suppose a polling company wants to estimate the percentage of voters in a city who support a particular candidate.
They take a random sample of 500 voters and find that 55% of them support the candidate. They want to calculate a 95% 
confidence interval for the true proportion of supporters in the population.

Using the formula for the margin of error, we get:

Margin of error = Z * (standard deviation / sqrt(sample size))
Z = 1.96 (for a 95% confidence level)
Standard deviation = sqrt(p * (1 - p) / n), where p is the sample proportion and n is the sample size
p = 0.55
n = 500

Plugging in the values, we get:

    Margin of error = 1.96 * (sqrt(0.55 * 0.45 / 500)) = 0.047

Therefore, the 95% confidence interval for the true proportion of supporters is 0.55 +/- 0.047, or 0.503 to 0.597. This means 
we can be 95% confident that the true proportion of supporters in the population falls within this range.

If the polling company were to increase their sample size to 1000, the margin of error would decrease, resulting in a smaller
range for the confidence interval and a more precise estimate of the true proportion of supporters in the population.
```

```python
Q10. In a study of the effectiveness of a new weight loss drug, a sample of 50 participants lost an average
of 6 pounds with a standard deviation of 2.5 pounds. Conduct a hypothesis test to determine if the drug is
significantly effective at a 95% confidence level using a t-test.
```

```python
  To determine if the new weight loss drug is significantly effective, we can use a one-sample t-test. The null hypothesis is 
    that the true population mean weight loss is equal to zero, while the alternative hypothesis is that the true population 
    mean weight loss is greater than zero. We will use a 95% confidence level, which corresponds to a significance level of 0.05.

The test statistic for a one-sample t-test is given by:

     t = (x̄ - μ) / (s / sqrt(n))

where x̄ is the sample mean, μ is the population mean (in this case, assumed to be zero), s is the sample standard deviation, 
and n is the sample size.

Plugging in the values from the problem, we get:

     t = (6 - 0) / (2.5 / sqrt(50))
    t = 13.42

The degrees of freedom for this test are (n - 1) = 49.

Using a t-distribution table or a calculator, we find the critical t-value for a one-tailed test with 49 degrees of freedom at
a significance level of 0.05 to be 1.676.

Since our calculated t-value (13.42) is greater than the critical t-value (1.676), we can reject the null hypothesis and 
conclude that the new weight loss drug is significantly effective at a 95% confidence level.
```

```python
 To conduct a hypothesis test to determine if the two teaching methods have a significant difference in student performance,
    we can use a two-sample t-test. The null hypothesis is that there is no significant difference between the two teaching 
    methods, while the alternative hypothesis is that there is a significant difference.

       H0: μA = μB
        Ha: μA ≠ μB

Where μA is the population mean score for sample A and μB is the population mean score for sample B.

Assuming equal variances, we can calculate the pooled standard deviation as follows:

Sp = sqrt(((nA-1)*sA^2 + (nB-1)*sB^2) / (nA + nB - 2))
where nA = nB = sample sizes of A and B
sA = standard deviation of sample A
sB = standard deviation of sample B

Plugging in the values, we get:

Sp = sqrt(((20-1)*6^2 + (20-1)*5^2) / (20+20-2)) = 5.527

Now, we can calculate the t-statistic using the following formula:

t = (x̄A - x̄B) / (Sp * sqrt(1/nA + 1/nB))
where x̄A = sample mean score for sample A
x̄B = sample mean score for sample B

Plugging in the values, we get:

t = (85 - 82) / (5.527 * sqrt(1/20 + 1/20)) = 3.25

Using a t-table with degrees of freedom of 38 (nA + nB - 2), we find that the critical t-value for a two-tailed test with a
significance level of 0.01 is approximately 2.704.

Since our calculated t-value of 3.25 is greater than the critical t-value of 2.704, we reject the null hypothesis and conclude
that there is a significant difference between the two teaching methods in terms of student performance.

In other words, we can say that there is strong evidence to suggest that one of the teaching methods is better than the other
in terms of student performance.
```

```python
Q14. In a study of the effects of caffeine on reaction time, a sample of 30 participants had an average
reaction time of 0.25 seconds with a standard deviation of 0.05 seconds. Conduct a hypothesis test to
determine if the caffeine has a significant effect on reaction time at a 90% confidence level using a t-test.
```

```python
  To conduct a hypothesis test, we need to state our null and alternative hypotheses:

Null hypothesis: The caffeine has no significant effect on reaction time. The population mean reaction time is equal to or
greater than 0.25 seconds.
Alternative hypothesis: The caffeine has a significant effect on reaction time. The population mean reaction time is less than
0.25 seconds.
We will use a one-tailed t-test with a significance level of 0.10 (90% confidence level) and a sample size of 30. The degrees 
of freedom are 29 (sample size minus 1).

We can calculate the t-statistic as:

t = (sample mean - hypothesized population mean) / (standard deviation / sqrt(sample size))

t = (0.25 - 0) / (0.05 / sqrt(30)) = 6.708

Using a t-distribution table or calculator with 29 degrees of freedom and a one-tailed test with a significance level of 0.10,
we find the critical value to be -1.311.

Since the calculated t-statistic (6.708) is greater than the critical value (-1.311), we reject the null hypothesis and 
conclude that the caffeine has a significant effect on reaction time at a 90% confidence level.

Therefore, we can say that there is strong evidence to suggest that caffeine has a positive effect on reaction time, making 
individuals more alert and responsive.
```
